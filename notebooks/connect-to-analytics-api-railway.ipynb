```python
from faker import Faker
import random
import requests
from pprint import pprint 
import os
import httpx
```

```python
api_base_url = 'http://' + os.environ.get('ANALYTICS_ENDPOINT') + ':8080' + '/' # 'https://fastapianalyticsmicroservice-production.up.railway.app/' # 'http://localhost:8002/'
api_path = 'api/events/'
endpoint = f'{api_base_url}{api_path}'
```

```python
print(endpoint)
```

```python
events_nb = 10_000
pages = [
    '/', '/about', '/pricing', '/contact', 
    '/blog', '/products', '/login', '/register',
    '/services', '/support', '/dashboard', '/settings'
]
```

```python
fake = Faker()
session_ids = [fake.uuid4() for _ in range(50)]

referrers = [
    'https://www.google.com',
    'https://www.bing.com',
    'https://www.yahoo.com',
    'https://www.duckduckgo.com',
    'https://www.ask.com',
    'https://www.brave.com',
    'https://www.yandex.com',
    'https://www.baidu.com',
    'https://www.linkedin.com',
    'https://www.facebook.com',
    'https://www.twitter.com',
    'https://www.instagram.com',
    'https://www.pinterest.com',
    'https://www.reddit.com',
    'https://www.tumblr.com',
    'https://www.quora.com',
    'https://www.wikipedia.org',
    'https://www.youtube.com',
    'https://www.twitch.tv',
    'https://www.tiktok.com',
    'https://www.snapchat.com',
    "" # Direct Traffic
]

user_agents = [
    fake.chrome(),
    fake.firefox(),
    fake.safari(),
    fake.chrome(),
    fake.user_agent(),
    fake.opera(),
    fake.internet_explorer()
]
```

```python
for _ in range(events_nb):
    page = random.choice(pages)
    referrer = random.choice(referrers)
    user_agent = random.choice(user_agents)
    session_id = random.choice(session_ids)

    payload = dict(
        page=page,
        user_agent=user_agent,
        ip_address=fake.ipv4(),
        session_id=session_id,
        referrer=referrer,
        duration=random.randint(30, 8400),
        description=f"this is the event description {_}"
    )

    response = httpx.post(endpoint, json=payload)
    if response.is_success:
        data = response.json()
        print(data)
    else:
        print(response.text)
```
